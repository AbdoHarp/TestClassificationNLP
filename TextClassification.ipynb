```python
def get_book(url):
    respons = request.urlopen(url)
    raw = respons.read()
    test = raw.decode("utf-8-sig")
    return test
```

```python
bookName = []
for i in range(0, int(book)):
    name = input("plz enter book name")
    bookName.append(name)
```

```python
def label(bn):
    l1 =bn[0]
    l2 =bn[1]
    lit1 =[l1 for i in range(200)]
    lit2 =[l2 for i in range(200)]
    return(lit1+lit2)
```

```python
target = label(bookName)
```

```python
def clean_data(holdBook):
    for i in range(len(holdBook)):
        holdBook[i] = holdBook[i].lower()
        holdBook[i] = re.sub("\n", " ", holdBook[i])
        holdBook[i] = re.sub("\r", " ", holdBook[i])
        holdBook[i] = holdBook[i].replace("\d+", "")
        holdBook[i] = holdBook[i].split(" ")
    for i in range(len(holdBook)):
        while ("" in holdBook[i]):
            holdBook[i].remove('')


clean_data(holdBook)
```

```python
chunk_size =200
def chunck(target):
    A = []
    for i in range(len(target)):
        chunckList= list()
        for a in range(0,len(target[i]), chunk_size):
            t=""
            for word in target[i][a:a+chunk_size]:
                t+=word+" "
            chunckList.append(t)
        A.append(chunckList)
    return A
```
